```python
# Get analytical soln
Nx = 256 + 1
Lx = 1.0
x = np.linspace(0, Lx, Nx)
t = np.linspace(t0, t_end+t0, T_eps2dx_Nx_256.shape[1])
```

```python
T_a = np.zeros([(T_eps2dx_Nx_256.shape[0]),(T_eps2dx_Nx_256.shape[1])])
i=0
while i<len(h_a): 
    h_now = h_a[i]
    mask = x < h_now
    T_a[mask,i] = 1.0 - erf(lambd*x[mask]/h_now)/erf(lambd)
    i = i+1
```
