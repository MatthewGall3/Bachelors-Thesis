```python
def radius(phi):
    idx = 0
    radius = np.zeros(101)

    while idx < 101:
    
        total_sum = np.sum(phi[:,:,idx])/(phi.shape[0]*phi.shape[1])
        radius[idx] = np.sqrt(total_sum/np.pi)
    
        idx += 1
    return radius
```

```python
Temperature_N256 = np.load('2D_Nx_256_eps_dx.npz')['T']
Temperature_N256.shape
```

```python
Nx = 256 + 1
Ny = 256 + 1
Lx = 1.0
Ly = 1.0
x = np.linspace(0, Lx, Nx)
y = np.linspace(0, Ly, Ny)
X, Y = np.meshgrid(x, y)
R = np.sqrt((X-0.5)**2 + (Y-0.5)**2)

t_idx = 10

R_flat = R.flatten()
T_flat = Temperature_N256[:,:,t_idx].flatten()

# Define radial bins
r_bins = np.linspace(0, np.max(R), 100)  # Adjust bin count if needed
r_centers = 0.5 * (r_bins[:-1] + r_bins[1:])  # Midpoints of bins

# Compute average temperature for each radial bin
T_radial = np.zeros(len(r_centers))

for i in range(len(r_centers)):
    mask = (R_flat >= r_bins[i]) & (R_flat < r_bins[i+1])  # Select points in bin
    T_radial[i] = np.mean(T_flat[mask])


# Plot Temperature vs. Radius
plt.figure(figsize=(8, 6))
plt.plot(r_centers, T_radial, 'o-', label=f'Temperature Profile at t_idx={t_idx}')
plt.xlabel("Radius")
plt.ylabel("Average Temperature")
plt.title(f"Radial Temperature Profile at t_idx={t_idx}")

plt.legend()
plt.show()
```

```python
def Temp_Calc(Temp, t_idx, Nx, Ny):
    
    x = np.linspace(0, Lx, Nx)
    y = np.linspace(0, Ly, Ny)
    X, Y = np.meshgrid(x, y)
    R = np.sqrt((X-0.5)**2 + (Y-0.5)**2)

    R_flat = R.flatten()
    T_flat = Temp[:,:,t_idx].flatten()

    r_bins = np.linspace(0, np.max(R), 100)  # Adjust bin count if needed
    r_centers = 0.5 * (r_bins[:-1] + r_bins[1:])  # Midpoints of bins

    T_radial = np.zeros(len(r_centers))

    for i in range(len(r_centers)):
        mask = (R_flat >= r_bins[i]) & (R_flat < r_bins[i+1])  # Select points in bin
        T_radial[i] = np.mean(T_flat[mask])

    return T_radial
```
